**Feature Engineering**


```python
#we will first remove the unwanted things from the dataframe
```

```python
df2=df.copy()
```

```python
df2.drop(["empid"],axis=1,inplace=True)
```

```python
df2.head()
```

```python
#now lets handle the missing values by using
df2.isnull().sum()
```

```python
#from above we can say that the satisfaction_level has 2 missing values
```

Filling the missing values by looking at stats!

```python
df2["satisfaction_level"].describe()
```

```python
#filling the missing values with the mean of thee satisfaction level using a method called fillna
df2["satisfaction_level"].fillna(df2["satisfaction_level"].mean(),inplace=True)
```

```python
df2.isnull().sum()
```

Handling the categorical feature in feature Engineering

```python
df2["salary"].unique()
```

```python
#to handle the categorical values we can use a method called getdummies
```

```python
salary_dummies=pd.get_dummies(df2["salary"],drop_first=True)
```

```python
salary_dummies
```

```python
df2=pd.concat([df2,salary_dummies],axis=1)
```

```python
df2
```

```python
df2.head()
```

```python
#droping the salary feature
df2.drop(["salary"],axis=1,inplace=True)
```
